```python
# Dependencies
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd
import requests
import pymongo
from flask import Flask, render_template, redirect
from flask_pymongo import PyMongo
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Setting up splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Mars News Site
url = 'https://redplanetscience.com/'
browser.visit(url)
```

```python
# Creating BeautifulSoup object; parsing with 'html.parser'
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Using the pretty() function to view how the tags are nested in the document
print(soup.prettify())
```

```python
# Identified from the web page the latest title
# From there, determined that the class is "content_title"

# Retrieving first title of article
# Using the get_text() function to return the text inside the Tag object 
news_title = soup.find_all("div", class_="content_title")[0].get_text()
print(news_title)
```

```python
# Identified from the web page the most recent paragraph text
# From there, determined that the class is "article_teaser_body"

# Retrieving first paragraph of the article
# Using the get_text() function to return the text inside the Tag object 
news_p = soup.find_all("div", class_="article_teaser_body")[0].get_text()
print(news_p)
```
